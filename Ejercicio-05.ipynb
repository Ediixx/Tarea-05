```python
import numpy as np

# Definir la funcin y su derivada
def funcion(x):
    return np.tan(np.pi * x) - 6

def derivada(x):
    return np.pi * (1 / np.cos(np.pi * x))**2

# Mtodo de Biseccin
def metodo_biseccion(funcion, a, b, tol=1e-6, max_iter=100):
    if funcion(a) * funcion(b) > 0:
        raise ValueError("La funcin no cambia de signo en el intervalo dado.")
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(funcion(c)) < tol:
            return c
        elif funcion(a) * funcion(c) < 0:
            b = c
        else:
            a = c
    return c

# Mtodo de Newton
def metodo_newton(funcion, derivada, p0, max_iter=10, tol=1e-6):
    p = p0
    for _ in range(max_iter):
        siguiente_punto = p - funcion(p) / derivada(p)
        if abs(siguiente_punto - p) < tol:
            return siguiente_punto
        p = siguiente_punto
    return p

# Mtodo de la Secante
def metodo_secante(funcion, p0, p1, tol=1e-6, max_iter=100):
    for _ in range(max_iter):
        if abs(p1 - p0) < tol:
            break
        p2 = p1 - funcion(p1) * (p1 - p0) / (funcion(p1) - funcion(p0))
        p0, p1 = p1, p2
    return p1

# Parmetros iniciales
punto_inicial_a = 0.48  # Punto inicial para el mtodo de Newton y la Secante
intervalo_b = (0.4, 0.5)  # Intervalo para el mtodo de Biseccin

# Clculo de la raz para cada mtodo

# Mtodo de Biseccin (literal a)
raiz_biseccion_a = metodo_biseccion(funcion, *intervalo_b, tol=1e-6)

# Mtodo de Newton (literal b)
raiz_newton_b = metodo_newton(funcion, derivada, punto_inicial_a, max_iter=10)

# Mtodo de la Secante (literal c)
raiz_secante_c = metodo_secante(funcion, punto_inicial_a, punto_inicial_a + 0.01, tol=1e-6)

# Resultados
print("Resultado del mtodo de Biseccin (literal a):", raiz_biseccion_a)
print("Resultado del mtodo de Newton (literal b):", raiz_newton_b)
print("Resultado del mtodo de la Secante (literal c):", raiz_secante_c)
```
