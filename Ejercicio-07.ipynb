```python
import numpy as np

# Función general para calcular la raíz con métodos específicos
def calcular_raiz(f, p0, p1=None, metodo='newton', derivada=None, tol=1e-6, max_iter=100):
    # Método de la Secante
    if metodo == 'secante' and p1 is not None:
        for _ in range(max_iter):
            p2 = p1 - f(p1) * (p1 - p0) / (f(p1) - f(p0))
            if abs(p2 - p1) < tol:
                return p2
            p0, p1 = p1, p2
        return p1
    
    # Método de Newton
    elif metodo == 'newton' and derivada is not None:
        for _ in range(max_iter):
            p_siguiente = p0 - f(p0) / derivada(p0)
            if abs(p_siguiente - p0) < tol:
                return p_siguiente
            p0 = p_siguiente
        return p0
    else:
        raise ValueError("Método o derivada no reconocidos")

# Función principal
def main():
    # Definir la función f(x)
    f = lambda x: x**(1/3)
    
    # Definir la derivada para el método de Newton
    derivada = lambda x: (1/3) * x**(-2/3)
    
    # Parámetros iniciales
    p0_secante = 5
    p1_secante = 0.5
    p0_newton = 1

    # Calcular la raíz usando el método de la secante
    raiz_secante = calcular_raiz(f, p0_secante, p1=p1_secante, metodo='secante')

    # Calcular la raíz usando el método de Newton
    raiz_newton = calcular_raiz(f, p0_newton, metodo='newton', derivada=derivada)

    # Imprimir resultados
    print("Raíz aproximada con el método de la Secante:", raiz_secante)
    print("Raíz aproximada con el método de Newton:", raiz_newton)

# Ejecutar el código
if __name__ == "__main__":
    main()
```
