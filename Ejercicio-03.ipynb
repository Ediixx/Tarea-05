```python
import numpy as np

# Definir las funciones para los problemas
def funcion_a(x):
    return 3 * x - np.exp(x)

def funcion_b(x):
    return 2 * x + 3 * np.cos(x) - np.exp(x)

# Derivadas de las funciones (para el método de Newton)
def derivada_a(x):
    return 3 - np.exp(x)

def derivada_b(x):
    return 2 - 3 * np.sin(x) - np.exp(x)

# Método de Newton
def metodo_newton(funcion, derivada, punto_inicial, tolerancia=1e-5, iteraciones_max=100):
    punto_actual = punto_inicial
    for _ in range(iteraciones_max):
        siguiente_punto = punto_actual - funcion(punto_actual) / derivada(punto_actual)
        if abs(siguiente_punto - punto_actual) < tolerancia:
            return siguiente_punto
        punto_actual = siguiente_punto
    return punto_actual

# Método de la Secante
def metodo_secante(funcion, inicio_1, inicio_2, tolerancia=1e-5, iteraciones_max=100):
    for _ in range(iteraciones_max):
        if abs(inicio_2 - inicio_1) < tolerancia:
            break
        siguiente = inicio_2 - funcion(inicio_2) * (inicio_2 - inicio_1) / (funcion(inicio_2) - funcion(inicio_1))
        if abs(siguiente - inicio_2) < tolerancia:
            return siguiente
        inicio_1, inicio_2 = inicio_2, siguiente
    return inicio_2

# Intervalos iniciales para los métodos
intervalo_a = (1, 2)
intervalo_b = (1, 2)

# Solución para el literal a
solucion_newton_a = metodo_newton(funcion_a, derivada_a, punto_inicial=1.5, tolerancia=1e-5)
solucion_secante_a = metodo_secante(funcion_a, *intervalo_a, tolerancia=1e-5)
print(f"Literal a: Solución con Newton = {solucion_newton_a}, Solución con Secante = {solucion_secante_a}")

# Solución para el literal b
solucion_newton_b = metodo_newton(funcion_b, derivada_b, punto_inicial=1.5, tolerancia=1e-5)
solucion_secante_b = metodo_secante(funcion_b, *intervalo_b, tolerancia=1e-5)
print(f"Literal b: Solución con Newton = {solucion_newton_b}, Solución con Secante = {solucion_secante_b}")
```
