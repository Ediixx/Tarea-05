```python
import numpy as np

# Definir la funcin y su derivada
def funcion(x):
    return 230 * x**4 + 18 * x**3 + 9 * x**2 - 221 * x - 9

def derivada(x):
    return 920 * x**3 + 54 * x**2 + 18 * x - 221

# Mtodo de la Secante
def metodo_secante(funcion, p0, p1, tolerancia=1e-6, max_iter=100):
    for _ in range(max_iter):
        if abs(p1 - p0) < tolerancia:
            break
        p2 = p1 - funcion(p1) * (p1 - p0) / (funcion(p1) - funcion(p0))
        if abs(p2 - p1) < tolerancia:
            return p2
        p0, p1 = p1, p2
    return p1

# Mtodo de Newton
def metodo_newton(funcion, derivada, p0, tolerancia=1e-6, max_iter=100):
    p = p0
    for _ in range(max_iter):
        siguiente_punto = p - funcion(p) / derivada(p)
        if abs(siguiente_punto - p) < tolerancia:
            return siguiente_punto
        p = siguiente_punto
    return p

# Intervalos para los ceros
intervalo_1 = (-1, 0)  # Intervalo [-1, 0]
intervalo_2 = (0, 1)   # Intervalo [0, 1]

# Clculo con el mtodo de la Secante (para el literal a y b)
# Literal a: Cero en el intervalo [-1, 0]
raiz_1_secante = metodo_secante(funcion, *intervalo_1, tolerancia=1e-6)

# Literal b: Cero en el intervalo [0, 1]
raiz_2_secante = metodo_secante(funcion, *intervalo_2, tolerancia=1e-6)

# Clculo con el mtodo de Newton (para el literal a y b)
punto_medio_1 = (intervalo_1[0] + intervalo_1[1]) / 2
punto_medio_2 = (intervalo_2[0] + intervalo_2[1]) / 2

# Literal a: Cero en el intervalo [-1, 0]
raiz_1_newton = metodo_newton(funcion, derivada, punto_medio_1, tolerancia=1e-6)

# Literal b: Cero en el intervalo [0, 1]
raiz_2_newton = metodo_newton(funcion, derivada, punto_medio_2, tolerancia=1e-6)

# Resultados
resultados = {
    "Secante": {"Cero en [-1, 0]": raiz_1_secante, "Cero en [0, 1]": raiz_2_secante},
    "Newton": {"Cero en [-1, 0]": raiz_1_newton, "Cero en [0, 1]": raiz_2_newton}
}

resultados
```
