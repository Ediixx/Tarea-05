```python
import numpy as np
from scipy.optimize import bisect
import matplotlib.pyplot as plt

# a. Definir la función f(x)
def f(x):
    return np.log(x**2 + 1) - np.exp(0.4 * x) * np.cos(np.pi * x)

# b. Graficar la función en un rango amplio para confirmar los ceros
def graficar_funcion(f, rango_x, titulo="Gráfica de la función"):
    x = np.linspace(rango_x[0], rango_x[1], 1000)
    y = f(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="f(x)")
    plt.axhline(0, color='red', linestyle='--', label="y = 0")
    plt.title(titulo)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid()
    plt.legend()
    plt.show()

# c. Graficar la función para verificar los ceros
graficar_funcion(f, rango_x=(0, 30), titulo="Gráfica de f(x) para encontrar los ceros")

# d. Determinar el intervalo inicial aproximado para el n-ésimo cero positivo
def intervalo_aproximado_n_eso(n):
    # Cada cero está aproximadamente separado por 1 unidad
    return (n - 1, n)

# Encontrar el vigesimoquinto cero positivo (n = 25)
n = 25
intervalo_aproximado = intervalo_aproximado_n_eso(n)
print(f"Intervalo inicial para el {n}-ésimo cero positivo: {intervalo_aproximado}")

# d. Usar el método de bisección para encontrar el cero
def encontrar_n_eso(f, intervalo, tol=1e-6):
    return bisect(f, *intervalo, xtol=tol)

# Calcular el vigesimoquinto cero positivo
cero_25 = encontrar_n_eso(f, intervalo_aproximado, tol=1e-6)
print(f"Vigesimoquinto cero positivo dentro de 10^-6: {cero_25}")
```
