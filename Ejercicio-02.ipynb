```python
import numpy as np

# Definir las funciones para cada caso
def funcion_a(x):
    return x**3 - 2*x**2 - 5

def funcion_b(x):
    return x**3 + 3*x**2 - 1

def funcion_c(x):
    return x - np.cos(x)

def funcion_d(x):
    return x - 0.8 - 0.2 * np.sin(x)

# Implementación del Método de la Secante
def metodo_secante(funcion, inicial_1, inicial_2, tolerancia=1e-4, iteraciones_max=100):
    for _ in range(iteraciones_max):
        if abs(inicial_2 - inicial_1) < tolerancia:
            break
        siguiente = inicial_2 - funcion(inicial_2) * (inicial_2 - inicial_1) / (funcion(inicial_2) - funcion(inicial_1))
        inicial_1, inicial_2 = inicial_2, siguiente
    return inicial_2

# Definir los intervalos iniciales para cada caso
intervalos = {
    "a": (1, 4),
    "b": (-3, -2),
    "c": (0, np.pi / 2),
    "d": (0, np.pi / 2)
}

# Caso a
solucion_a = metodo_secante(funcion_a, *intervalos["a"], tolerancia=1e-4)
print(f"Solución para el caso a: {solucion_a}")

# Caso b
solucion_b = metodo_secante(funcion_b, *intervalos["b"], tolerancia=1e-4)
print(f"Solución para el caso b: {solucion_b}")

# Caso c
solucion_c = metodo_secante(funcion_c, *intervalos["c"], tolerancia=1e-4)
print(f"Solución para el caso c: {solucion_c}")

# Caso d
solucion_d = metodo_secante(funcion_d, *intervalos["d"], tolerancia=1e-4)
print(f"Solución para el caso d: {solucion_d}")
```
