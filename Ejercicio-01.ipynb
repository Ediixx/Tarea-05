```python
import numpy as np

# Definición de la función f(x) y su derivada f'(x)
def funcion(x):
    return -x**3 - np.cos(x)

def derivada(x):
    return -3*x**2 + np.sin(x)

# Método de Newton
def metodo_newton(funcion, derivada, inicial, iteraciones_max=2):
    x = inicial
    for _ in range(iteraciones_max):
        x = x - funcion(x) / derivada(x)
    return x

# Método de la Secante
def metodo_secante(funcion, inicial_1, inicial_2, iteraciones_max=2):
    x0, x1 = inicial_1, inicial_2
    for _ in range(iteraciones_max):
        x = x1 - funcion(x1) * (x1 - x0) / (funcion(x1) - funcion(x0))
        x0, x1 = x1, x  # Actualizar los valores
    return x

# Parámetros iniciales
inicial_newton = -5
inicial_secante_1 = -7
inicial_secante_2 = -1.5

# Cálculos
resultado_newton = metodo_newton(funcion, derivada, inicial_newton)
resultado_secante = metodo_secante(funcion, inicial_secante_1, inicial_secante_2)

resultado_newton, resultado_secante
```
